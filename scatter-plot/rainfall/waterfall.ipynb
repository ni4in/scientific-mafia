```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python

df = pd.read_csv("rainfall.csv");
data = df[df.columns[1:4]].to_numpy()
labels = df[df.columns[0]].to_numpy()
fig,ax = plt.subplots(figsize = (40,30))
x = data[:,0]
y = data[:,1]
z = data[:,2]
t = z
xmin = x.min()-500 
xmax = x.max()+500
ax.set_xlim([xmin,xmax])
ax.set_ylim([0,100])
ax.set_yticks(np.arange(0,100,10))
ax.set_xticks(np.arange(xmin,xmax,500))
ax.scatter(x,y, s = z, color= 'green')
ax.grid()
plt.ylabel("Percentage Forest", fontsize = 20)
plt.xlabel("Annual Rainfall in mm", fontsize = 20)
plt.title("Percentage Forest vs Annual Rainfall vs Population Density", fontsize = 20)
for i, txt in enumerate(labels):
    ax.annotate(txt, (x[i], y[i]), fontsize = 20, textcoords = "offset points",ha = 'center', xytext=(0,10))

```
